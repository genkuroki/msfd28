````julia
using Plots
X = cumsum(randn(2^14,3), dims=1)
plotlyjs(legend=false)
plot(X[:,1], X[:,2], X[:,3], lw=0.4, size=(600, 500))
````

**注意:** 上のセルを正常に実行するためには [plotly-orca](https://github.com/plotly/orca) をインストールしてから, 

```julia
pkg> add PlotlyJS
pkg> add ORCA
pkg> build ORCA
```

しておく必要がある. JuliaのConda.jlで管理しているminiconda内に plotly-orca をインストールするには

```julia
using Conda
Conda.add_channel("plotly")
Conda.add("plotly-orca")
```

を実行すればよい. 詳しくは [https://github.com/plotly/orca](https://github.com/plotly/orca) を参照せよ.
