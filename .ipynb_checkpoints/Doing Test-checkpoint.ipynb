```julia
f = open(io->read(io, String), "test.jmd", "r")
f = replace(f, r"[^#]# ([^\n]*)\n"=>s"---
title: \1")
f = replace(f, r"\n\* Author:"=>s"
author:")
f = replace(f, r"\n\* Date: ([^\n]*)\n"=>s"
date: \1
---
")
f = replace(f, "\n#"=>"\n")
print(f)
open(io->write(io, f), "hoge.jmd", "w")
```

```julia
f = open(io->read(io, String), "hoge.tex", "r")
f = replace(f, "\n\\title{"=>"\n\\title{\\bfseries")
#f = replace(f, "\n\\author{"=>"\n\\author{\\bfseries")
f = replace(f, r"\n*\\\["=>"\n\\[")
f = replace(f, r"\n*\\begin{aligned}"=>"\n\\[\n\\begin{aligned}")
f = replace(f, r"\n\\end{aligned}\n*"=>"\n\\end{aligned}\n\\]\n")
f = replace(f, r"\\includegraphics\[width=\\linewidth\]{([^}]*)}"=>s"{\\centering\\includegraphics[width=0.8\\linewidth]{\1}}")
print(f)
```

```julia
open(io->write(io, f), "moge.tex", "w")
```
